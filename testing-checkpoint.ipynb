```python
font = {'family' : 'DejaVu Sans',
        'weight' : 'normal',
        'size'   : 16}

matplotlib.rc('font', **font)

flightpaths = pd.read_csv('TOflightpaths.txt',sep='\t')
city_names = ['Accra','Cotonou','Lomé','Savé']
city_lats = [5.55,6.36667,6.13194,8.03333]; city_lons = [-0.2,2.43333,1.22278,2.48333]
box_lats = [6.4,8.2,8.2,6.4,6.4]; box_lons = [1.3,1.3,2.8,2.8,1.3]

mapfig, ax = plt.subplots(subplot_kw={'projection': ccrs.PlateCarree()}, figsize=(20,10))

#ax.set_extent([-1,3, 4,8])

countriesname = 'ne_10m_admin_0_countries.shp'
lakesname = 'ne_10m_lakes.shp'
#oceansname = 'ne_10m_ocean.shp'
countries = ShapelyFeature(Reader(countriesname).geometries(),ccrs.PlateCarree(),facecolor='#E1D5AA')
lakes = ShapelyFeature(Reader(lakesname).geometries(),ccrs.PlateCarree(),facecolor='#C4E0F6')

#oceans = ShapelyFeature(Reader(oceansname).geometries(),ccrs.PlateCarree(),facecolor='#98C3CB')
ax.add_feature(countries, edgecolor='black', zorder=1)
ax.add_feature(lakes, edgecolor='black', zorder=1)
#ax.add_feature(oceans, edgecolor='None', zorder=1)
ax.background_patch.set_facecolor('#C4E0F6')

ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True, linewidth=1, color='black', alpha=0.5, linestyle='--')

ax.plot(flightpaths['TO_lon'], flightpaths['TO_lat'], transform=ccrs.PlateCarree(), color='#AC434B', zorder=2)
ax.scatter(city_lons, city_lats, transform=ccrs.PlateCarree(), marker='*', color='blue', s=300, zorder=3)
ax.plot(box_lons, box_lats, transform=ccrs.PlateCarree(), color='#242AFB')

ax.text(city_lons[0], city_lats[0]-0.14, city_names[0], horizontalalignment='center', transform=ccrs.PlateCarree()) # Accra
ax.text(city_lons[1], city_lats[1]-0.15, city_names[1], horizontalalignment='center', transform=ccrs.PlateCarree()) # Cotonou
ax.text(city_lons[2]+0.24, city_lats[2]-0.12, city_names[2], horizontalalignment='center', transform=ccrs.PlateCarree()) # Lomé
ax.text(city_lons[3]+0.05, city_lats[3]+0.07, city_names[3], horizontalalignment='center', transform=ccrs.PlateCarree()) # Savé
ax.text(-0.08, 8.24, 'Ghana', horizontalalignment='center', transform=ccrs.PlateCarree(), style='italic')
ax.text(1.15, 8.24, 'Togo', horizontalalignment='center', transform=ccrs.PlateCarree(), style='italic')
ax.text(2.2, 8.24, 'Benin', horizontalalignment='center', transform=ccrs.PlateCarree(), style='italic')
```
